```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
dataset = pd.read_csv('dataset_1.csv')
```

```python
dataset.head(10)
```

```python
dataset.info()
```

## Category

```python
dataset['category'] = dataset['category'].replace("cereals and tubers", "cereals_and_tubers")
dataset['category'] = dataset['category'].replace("miscellaneous food", "miscellaneous_food")
dataset['category'] = dataset['category'].replace("oil and fats", "oil_and_fats")
dataset['category'] = dataset['category'].replace("pulses and nuts", "pulses_and_nuts")
dataset['category'] = dataset['category'].replace("vegetables and fruits", "vegetables_and_fruits")
```

## Unit 

```python
dataset['unit'] = dataset['unit'].replace("KG", "0")
dataset['unit'] = dataset['unit'].replace("100 KG", "1")
dataset['unit'] = dataset['unit'].replace("L", "2")
```

## Pricetype

```python
dataset['pricetype'] = dataset['pricetype'].replace("Retail", "1")
dataset['pricetype'] = dataset['pricetype'].replace("Wholesale", "0")
```
